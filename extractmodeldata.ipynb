```python
#keep
def getfilenamelist(folderpath):
    filenames = os.listdir(folderpath)
    filenamelist = []
    for i in filenames:
        if i[:1] != '.':
            filenamelist.append(i)
    return filenamelist
```

```python
def searchmodel(filenamelist, modelname):
    targetfiles = []
    for logfile in filenamelist:
        with open(folderpath + logfile, 'r') as currentlogfile:
            currentline = currentlogfile.readline()
            if currentline != '':
                for i in range(len(logfile)):
                    if currentline.find(modelname) != -1:
                        targetfiles.append(logfile)
                        break
                    else:
                        currentline = currentlogfile.readline()
    return targetfiles
```

```python
#utility
def gettimestamp(line):
    timestamp_location = string.find(line,"timestamp")
    return line[timestamp_location+11:timestamp_location+24]
```
