```python
import os
import re
import csv
import string
import datetime
import linecache

folderpath = '/Users/ygp072/Documents/ModelSim/data analysis/modelsim_log_2014_test/'

def parselogfolder(folderpath, modelfeature, outputcsvname, outputtype):
    filesinfolder = getfilesinfolder(folderpath)
    targetlogs = gettargetlogs(filesinfolder,modelfeature)
    parselogs(targetlogs, outputcsvname, outputtype)
    
def getfilesinfolder(folderpath):
    filenames = os.listdir(folderpath)
    filesinfolder = []
    for i in filenames:
        if i[:1] != '.':
            filesinfolder.append(folderpath + i)
    return filesinfolder

def gettargetlogs(filesinfolder,modelfeature):
    targetlogs = []
    for logfile in filesinfolder:
        with open(logfile, 'r') as currentlogfile:
            currentline = currentlogfile.readline()
            if currentline != '':
                for i in range(len(logfile)):
                    if currentline.find(modelfeature) != -1:
                        targetlogs.append(logfile)
                        break
                    else:
                        currentline = currentlogfile.readline()
    return targetlogs

def gettimestamp(line):
    timestamp_location = string.find(line,"timestamp")
    return line[timestamp_location+11:timestamp_location+24]

def getsliderbriefs(slider_events):
    if len(slider_events) > 0:
        slider_briefs = []
        slider_events_range = []
        slider_event_start = int(slider_events[0][2])

        for i in range(len(slider_events) - 1):
            if slider_events[i][1] != slider_events[i+1][1]:
                slider_events_range.append([slider_event_start, int(slider_events[i][2])])
                slider_event_start = int(slider_events[i+1][2])
        slider_events_range.append([slider_event_start, int(slider_events[len(slider_events)-1][2])])

        for i in slider_events_range:
            for j in slider_events:
                if i[0] == j[2] :
                    slider_briefs.append([j[0],j[1],i[0],i[1]])#content[slider_events_range[i][0] + 3][11:-9],content[slider_events_range[i][1] + 3][11:-9]]) 
    
        return sorted(slider_briefs)
    else:
        return []

def getbuttonbriefs(button_list):
    if len(button_list) > 0:
        cleanbuttonlist = []
        sortedbuttonlist = sorted(button_list)
        for i in sortedbuttonlist:
            if i[1][:8]=='go/pause':
                cleanbuttonlist.append(i)
            elif i[1][-7:]=='pressed':
                cleanbuttonlist.append(i)
    return sorted(cleanbuttonlist)

def getmessagebriefs(message_list):
    processed_list = []
    processed_list = message_list
    return processed_list
    
    
def parselogs(targetlogs, outputcsvname, outputtype):
    with open(outputcsvname, 'wb') as f:
        writer = csv.writer(f)
        
        for log in targetlogs:
            actionstartlinenum = 0
            slider_event_list = []
            slider_linenum_list = []
            button_event_list = []
            message_event_list = []
            rawlist = []
            brieflist = []
                
            with open(log, 'r') as currentlog:#currentlog as file obj
                content = currentlog.readlines()
                currentlog.close()
                
                for i in content:
            
                    linenum = content.index(i)
                    if i.find('username=') != -1:
                        username = re.search("username=\".+\" n", i).group(0)[10:-3]
                        name = re.search(" name=\"(.)*\" i", i).group(0)[7:-3]
                        modelname = re.search(" modelName=\"(.)*\" v", i).group(0)[12:-3]
            
                    if i.find('<error') != -1:
                        actionstartlinenum = content.index(i)
                    
                    if actionstartlinenum != 0 :#and linenum < len(content) - 2  
            
                        if "type=\"button\"" in i and linenum > actionstartlinenum :
                            button_time = gettimestamp(i)
                            button_name = content[linenum+1][10:-8]
                        
                            if "pressed" in content[linenum+2]:
                                button_event_list.append([button_time,button_name + " pressed"])
                                    
                            if "released" in content[linenum+2]:
                                button_event_list.append([button_time,button_name + " released"])
                                
                        if "type=\"custom message\"" in i and linenum > actionstartlinenum :
                            message_time = gettimestamp(i)
                            message_event_list.append([message_time,content[linenum+1][13:-11]])
                            
                        if "type=\"slider\"" in i and linenum > actionstartlinenum :
                            slider_time = gettimestamp(i)
                            slider_name = content[linenum+2][10:-8]
                            slider_event_list.append([slider_time,content[linenum+2][10:-8]])
                            slider_linenum_list.append([slider_time,content[linenum+2][10:-8],linenum])
            
                sliderbriefs = getsliderbriefs(sorted(slider_linenum_list))
                buttonbriefs = getbuttonbriefs(sorted(button_event_list))
                messagebriefs = getmessagebriefs(sorted(message_event_list))
                
                rawhalflist = sorted(slider_event_list + button_event_list + message_event_list)
                briefhalflist = sorted(sliderbriefs + buttonbriefs + messagebriefs)
               
                for i in rawhalflist:
                    rawlist.append([username,name,modelname,i[0],i[1]])
                
                for i in briefhalflist:
                    if len(i) > 2:
                        brieflist.append([username,name,modelname,i[0],i[1],content[i[2]+3][11:-9],content[i[3]+3][11:-9]])
                    else:
                        brieflist.append([username,name,modelname,i[0],i[1]])
                        
                if outputtype == 'raw':
                    writer.writerows(rawlist)
                elif outputtype == 'brief':
                    writer.writerows(brieflist)
                else:
                    print ('OutPutType must be either \'brief\' or \'raw\'')  
    f.close()    
```
